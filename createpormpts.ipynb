```python
from datasets import load_from_disk

# 1. Load the dataset
ds = load_from_disk('./few-nerd_train')

# 2. Get the label names from the features
coarse_labels = ds.features['ner_tags'].feature.names
fine_labels = ds.features['fine_ner_tags'].feature.names

# 3. Pick an example (e.g., the first one)
example = ds[1]

# 4. Map the integer tags to their string names and display
print(f"{'Token':<20} | {'Coarse Tag':<15} | {'Fine Tag'}")
print("-" * 60)

for token, coarse_id, fine_id in zip(example['tokens'], example['ner_tags'], example['fine_ner_tags']):
    coarse_name = coarse_labels[coarse_id]
    fine_name = fine_labels[fine_id]
    
    # Only color/highlight if it's not 'O' (optional visual aid)
    label_display = f"{coarse_name:<15} | {fine_name}"
    
    print(f"{token:<20} | {label_display}")
```
